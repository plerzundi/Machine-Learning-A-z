```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
# Importación del Datasets
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset.head()
```

```python
# Separar variable independiente y dependiente
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1:].values
```

```python
# Codificar los datos categoricos de las ciudades
ct = ColumnTransformer([('one_hot_encoder',OneHotEncoder(categories='auto'),[3])],remainder='passthrough')
X =np.array(ct.fit_transform(X),dtype=np.float64)
```

```python
# Evitar trampas de las variables dummy ficticias aplicar n-1
X = X[:,1:]
```

```python
# Dividir el conjunto de entrenamiento con los de testing
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Ajustar el modelo de regresion lineal multiplo con el conjunto de entrenamiento
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
# Prediccion de los conjuntos de datos de testing
y_pred = regression.predict(X_test)
```

```python
# Contruir un modelo óptimo de RLM utilizando la eliminación hacia atrás
X = np.append(arr = np.ones((50,1)).astype(int), values=X,axis=1 ) # axis=1 columna, axis=0 fila
SL = 0.05 # nivel de significacion
X_opt = X[:,[0,1,2,3,4,5]] # va guardar las variables optimas
regression_OLS = sm.OLS(endog=y,exog=X_opt.tolist()).fit()
regression_OLS.summary()


```
